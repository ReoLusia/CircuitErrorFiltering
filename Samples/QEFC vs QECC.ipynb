```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 1, 0.01)

fid_Q1 = 1 - x
fid_Q2 = 1 - 3*pow(x,2)*(1-x) - pow(x,3)
fid_Q3 = pow(1-x,2) / (pow(1-x,2) + pow(x,2))

# plotting the points
plt.plot(x, fid_Q1, label="$Q_1$") 
plt.plot(x, fid_Q2, label="$Q_2$ (QEFC)")
plt.plot(x, fid_Q3, label="$Q_3$ (QECC)")
  
# function to show the plot
plt.xlabel("$p_e$")
plt.ylabel("Fidelity")
plt.xlim([0, 1/2]) 
plt.ylim([1/2, 1]) 
plt.grid(True, axis='y', alpha=0.5, linestyle='--')
plt.legend(loc='best')
plt.show()
```

```python
prob = pow(1-x,2) + pow(x,2)

plt.plot(x, prob, color='black') 
# function to show the plot
plt.xlabel("$p_e$")
plt.ylabel("$p_s$")
plt.xlim([0, 1/2]) 
plt.ylim([1/2, 1]) 
plt.grid(True, axis='y', alpha=0.5, linestyle='--')
#plt.legend(loc='best')
plt.show()
```
